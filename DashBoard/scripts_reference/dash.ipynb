```python
#gerador de csv - dados demograficos
import pandas as pd
import numpy as np

# Dados básicos
estados = [
    'Acre', 'Alagoas', 'Amapá', 'Amazonas', 'Bahia', 'Ceará', 'Distrito Federal',
    'Espírito Santo', 'Goiás', 'Maranhão', 'Mato Grosso', 'Mato Grosso do Sul',
    'Minas Gerais', 'Pará', 'Paraíba', 'Paraná', 'Pernambuco', 'Piauí', 'Rio de Janeiro',
    'Rio Grande do Norte', 'Rio Grande do Sul', 'Rondônia', 'Roraima', 'Santa Catarina',
    'São Paulo', 'Sergipe', 'Tocantins'
]

regioes = {
    'Norte': ['Acre', 'Amapá', 'Amazonas', 'Pará', 'Rondônia', 'Roraima', 'Tocantins'],
    'Nordeste': ['Alagoas', 'Bahia', 'Ceará', 'Maranhão', 'Paraíba', 'Pernambuco', 
                 'Piauí', 'Rio Grande do Norte', 'Sergipe'],
    'Centro-Oeste': ['Distrito Federal', 'Goiás', 'Mato Grosso', 'Mato Grosso do Sul'],
    'Sudeste': ['Espírito Santo', 'Minas Gerais', 'Rio de Janeiro', 'São Paulo'],
    'Sul': ['Paraná', 'Rio Grande do Sul', 'Santa Catarina']
}

# Função para pegar a região
def get_regiao(estado):
    for regiao, lista in regioes.items():
        if estado in lista:
            return regiao
    return None

# Criação do DataFrame
dados = []

for estado in estados:
    populacao_total = np.random.randint(400_000, 46_000_000)
    populacao_masculina = int(populacao_total * np.random.uniform(0.48, 0.51))
    populacao_feminina = populacao_total - populacao_masculina
    crescimento_populacional = round(np.random.uniform(0.5, 1.8), 2)
    natalidade = round(np.random.uniform(12.0, 22.0), 1)
    mortalidade = round(np.random.uniform(4.5, 8.0), 1)
    expectativa_vida = round(np.random.uniform(68.0, 79.0), 1)
    idh = round(np.random.uniform(0.65, 0.87), 3)
    pib_per_capita = round(np.random.uniform(18000, 80000), 2)
    desemprego = round(np.random.uniform(5.5, 14.5), 1)
    indice_saneamento = round(np.random.uniform(40.0, 98.0), 1)
    acesso_saude_publica = round(np.random.uniform(70.0, 99.0), 1)
    
    dados.append({
        "estado": estado,
        "regiao": get_regiao(estado),
        "populacao_total": populacao_total,
        "populacao_masculina": populacao_masculina,
        "populacao_feminina": populacao_feminina,
        "crescimento_populacional": crescimento_populacional,
        "natalidade": natalidade,
        "mortalidade": mortalidade,
        "expectativa_vida": expectativa_vida,
        "idh": idh,
        "pib_per_capita": pib_per_capita,
        "desemprego": desemprego,
        "indice_saneamento": indice_saneamento,
        "acesso_saude_publica": acesso_saude_publica
    })

df = pd.DataFrame(dados)

# Salvar em CSV
df.to_csv("dados_demograficos_brasil.csv", index=False, sep=";")

# Ou salvar em Excel
df.to_excel("dados_demograficos_brasil.xlsx", index=False)

print("Arquivo CSV e Excel gerados com sucesso!")
print(df.head())
```

```python
#grafico do mapa do brasil com uso da planilha
import vizro.models as vm
import vizro.plotly.express as px
from vizro import Vizro
import pandas as pd
from vizro.tables import dash_ag_grid
from config import SIGLAS_ESTADOS, REGIOES_ESTADOS

def carregar_dados_planilha(caminho_arquivo):
    """Carrega os dados da planilha manual"""
    df = pd.read_excel(caminho_arquivo, header=None, usecols=[0, 1], names=['estado', 'populacao'])
    
    # Adiciona siglas e regiões
    df['sigla'] = df['estado'].map(SIGLAS_ESTADOS)
    df['regiao'] = df['sigla'].map(REGIOES_ESTADOS)
    
    return df.dropna()

# Carrega os dados
df = carregar_dados_planilha("POP2022_Brasil_e_UFs.xlsx")

# Cria o mapa coroplético sem título interno
fig = px.choropleth(
    df,
    geojson="https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/brazil-states.geojson",
    locations='sigla',
    featureidkey="properties.sigla",
    color='populacao',
    hover_name='estado',
    hover_data={'regiao': True, 'populacao': ':.1f'},
    color_continuous_scale='Blues',
    scope='south america',
    labels={'populacao': 'População (milhares)'}
)

# Configurações do layout
fig.update_geos(visible=False, showsubunits=True, subunitcolor='gray', fitbounds="locations")
fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0})

# Prepara dados para as tabelas
df_resumo = df.groupby('regiao').agg(
    total_populacao=('populacao', 'sum'),
    num_estados=('estado', 'count')
).reset_index()
df_resumo['total_populacao'] = df_resumo['total_populacao'].round(1)

df_detalhado = df[['regiao', 'estado', 'populacao']].sort_values(['regiao', 'populacao'], ascending=[True, False])

# Função para criar tabelas sem o parâmetro title
def criar_tabela_ag_grid(data_frame):
    return dash_ag_grid(data_frame=data_frame)

# Cria páginas do dashboard
pages = [
    vm.Page(
        title="Visão Nacional",
        layout=vm.Grid(grid=[[0, 1], [2, 2]]),
        components=[
            # Mapa
            vm.Graph(
                figure=fig,
                title="Mapa de Distribuição Populacional"  # Título definido no componente
            ),
            # Tabela resumo
            vm.AgGrid(
                figure=criar_tabela_ag_grid(
                    data_frame=df_resumo.sort_values('total_populacao', ascending=False)
                ),
                title="Dados por Região"  # Título definido no componente
            ),
            # Tabela detalhada
            vm.AgGrid(
                figure=criar_tabela_ag_grid(
                    data_frame=df_detalhado
                ),
                title="Dados por Estado"  # Título definido no componente
            )
        ],
        controls=[
            vm.Filter(column="regiao")
        ]
    )
]

# Adiciona páginas por região
for regiao in sorted(df['regiao'].unique()):
    df_regiao = df[df['regiao'] == regiao]
    
    # Cria mapa regional sem título interno
    fig_regiao = px.choropleth(
        df_regiao,
        geojson=fig.data[0].geojson,
        locations='sigla',
        featureidkey="properties.sigla",
        color='populacao',
        scope='south america'
    )
    
    pages.append(
        vm.Page(
            title=regiao,
            layout=vm.Grid(grid=[[0, 1]]),
            components=[
                vm.Graph(
                    figure=fig_regiao,
                    title=f"Mapa da {regiao}"  # Título definido no componente
                ),
                vm.AgGrid(
                    figure=criar_tabela_ag_grid(
                        data_frame=df_regiao[['estado', 'populacao']].sort_values('populacao', ascending=False)
                    ),
                    title="Dados Detalhados"  # Título definido no componente
                )
            ]
        )
    )

# Cria e executa o dashboard
dashboard = vm.Dashboard(pages=pages)
Vizro().build(dashboard).run()
```

```python
#versao 3.0 (atual)
import pandas as pd
import vizro.plotly.express as px
from vizro import Vizro
import vizro.models as vm
from vizro.tables import dash_ag_grid
from config import SIGLAS_ESTADOS, REGIOES_ESTADOS

Vizro._reset()

def carregar_dados_planilha(caminho_arquivo):
    df2 = pd.read_excel(caminho_arquivo, header=None, usecols=[0, 1], names=['estado', 'populacao'])  
    df2['sigla'] = df2['estado'].map(SIGLAS_ESTADOS)
    df2['regiao'] = df2['sigla'].map(REGIOES_ESTADOS)
    return df2.dropna()

#CSV
df = pd.read_csv("base/dados_demograficos_brasil.csv", sep=";")
df1 = pd.read_csv("base/database_bioscan.csv", sep=";")
df2 = carregar_dados_planilha("base/POP2022_Brasil_e_UFs.xlsx")

# Cria o mapa coroplético
fig = px.choropleth(
    df2,
    geojson="https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/brazil-states.geojson",
    locations='sigla',
    featureidkey="properties.sigla",
    color='populacao',
    hover_name='estado',
    hover_data={'regiao': True, 'populacao': ':.1f'},
    color_continuous_scale='Blues',
    scope='south america',
    labels={'populacao': 'População (M)'}
)

#layout mapa
fig.update_geos(visible=False, showsubunits=True, subunitcolor='gray', fitbounds="locations")
fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0})

#dados das tabelas
df_resumo = df2.groupby('regiao').agg(total_populacao=('populacao', 'sum'), num_estados=('estado', 'count')).reset_index()
df_resumo['total_populacao'] = df_resumo['total_populacao'].round(1)
df_detalhado = df2[['regiao', 'estado', 'populacao']].sort_values(['regiao', 'populacao'], ascending=[True, False])

#tabelas sem o parâmetro title
def criar_tabela_ag_grid(data_frame):
    return dash_ag_grid(data_frame=data_frame)

#calculos para graficos
total_masculino = df["populacao_masculina"].sum()
total_feminino = df["populacao_feminina"].sum()
df_donut = pd.DataFrame({
    "Categoria": ["População Masculina", "População Feminina"],
    "Total": [total_masculino, total_feminino]
})

# Preparação de figuras 
fig_pie = px.pie(df_donut, values="Total", names="Categoria", hole=0.5)
fig_crescimento = px.histogram(df, x='regiao', y='crescimento_populacional')

#################################################################################################################

# Definição das telas dashboard
page0 = vm.Page(
    title="Homepage",
    id="main_page",
    layout=vm.Grid(grid=[[0,1,2],[0,3,4]]),
    components=[
        vm.Card(
            text="""
            ## Lumina BI
            Lumina Lab desenvolveu uma plataforma completa para monitoramento de dados, 
            oferecendo dashboards modernos e altamente interativos. Acesse os KPI's dos nossos sistemas!
            
            Esses recursos foram criados para facilitar a rotina de gestores de projetos, 
            permitindo:

            -> Visualização de indicadores estratégicos

            -> Análise de desempenho em tempo real

            -> Identificação de riscos e oportunidades

            -> Tomada de decisão baseada em dados confiáveis

            Com foco em tecnologia, segurança e usabilidade, transformamos dados em insights 
            claros e otimizados que potencializem resultados.
            
            ## Acesse ao lado o dashboard dos nossos produtos Lumina Lab!
            
            """,
            extra={"style": {"height": "600px", "width": "400px"}},
        ),
        vm.Card(
            text="""
            ### 
            ![](assets/images/denguezero.png#my-bioscan)
            """,
            href="/bioscan_dashboard_v3",
        ),
        vm.Card(
            text="""
            ![](assets/images/empoli_2.png#my-empoli)
            """,
            href="/dados_empoli",
        ),
        vm.Card(
            text="""
            ###
            ![](assets/images/_01_urban.png#my-urban)
            """,
            href="/urbanai",
        ),
        vm.Card(
            text="""
            ###
            ![](assets/images/eleodora2.png#my-eleodora)
            """,
            href="/dados_gerais",
        ),
    ],
)
      
    
#################################################################################################################
#  Telas de Direcionamentos - paginas

page1 = vm.Page(
    title="Bioscan - Dengue Zero",
    id="bioscan_dashboard_v3",
    layout=vm.Grid(grid=[[0, 1],[2, 3]]),
    components=[
        vm.Container(
            title="População por Sexo",
            components=[vm.Graph(
                figure=fig_pie,
            )],
            variant="filled"
        ),
        vm.Container(
            title="Acesso à Saúde Publica",
            components=[vm.Graph(
                figure=px.histogram(df, x="regiao", y="acesso_saude_publica"),
            )],
            variant="filled"
        ),
        vm.Container(
            title="Combate à Dengue",
            components=[vm.Graph(
            figure=px.scatter(df1, x="estados", y="focos_encontrados")
            )],
            variant="filled"
        ),
        vm.Container(
            title="Indice de Saneamento",
            components=[vm.Graph(
                figure=px.histogram(df, x="regiao", y="indice_saneamento"),
            )],
            variant="filled"
        ), 
    ],
)

page2 = vm.Page(
    title="Empoli",
    id="dados_empoli",
    layout=vm.Grid(grid=[[0, 1],[2, 3]]),
    components=[
        vm.Container(
            title="Expectativa de Vida por Região",
            components=[vm.Graph(
                figure=px.histogram(df, y="regiao", x="expectativa_vida"),
            )],
            variant="filled",
        ),
        vm.Container(
            title="IDH por Região",
            components=[vm.Graph(
                figure=px.histogram(df, y="regiao", x="idh"),
            )],
            variant="filled",
        ),
        vm.Container(
            title="Distribuição Populacional",
            components=[vm.Graph(
                figure=px.histogram(df, x="estado", y="populacao_total"),
            )],
            variant="filled",
        ),
        vm.Container(
            title="Crescimento Populacional por Região",
            components=[vm.Graph(
                figure=fig_crescimento,
            )],
            variant="filled"
        ),
    ],
    controls=[vm.Filter(column="regiao")]
)

page3 = vm.Page(
    title="Eleodora",
    id="dados_gerais",
    components=[
        vm.AgGrid(
            figure=dash_ag_grid(data_frame=df1)),
    ],
)
page4 = vm.Page(
    title="Urban AI",
    id="urbanai",
    layout=vm.Grid(grid=[[0, 1], [2, 2]]),
        components=[
            vm.Graph(
                figure=fig, title="Mapa de Distribuição Populacional"),
            vm.AgGrid(
                figure=criar_tabela_ag_grid(data_frame=df_resumo.sort_values('total_populacao', ascending=False)),
                title="Dados por Região"),
            vm.AgGrid(
                figure=criar_tabela_ag_grid(data_frame=df_detalhado),
                title="Dados por Estado")
        ],
        controls=[
            vm.Filter(column="regiao")
        ],
    )

#dashboard = vm.Dashboard(pages=[page0, page1, page2, page3, page4],  
                         #navigation=vm.Navigation(pages=["main_page"]))

dashboard = vm.Dashboard(
    pages=[page0, page1, page2, page3, page4],
    navigation=vm.Navigation(
        nav_selector=vm.NavBar(
            items=[
                vm.NavLink(
                    label="Homepage", icon="Apps", pages=["main_page"],),
                vm.NavLink(
                    label="Dengue Zero", icon="Lab Research", pages=["bioscan_dashboard_v3"],),
                vm.NavLink(
                    label="Empoli", icon="Gite", pages=["dados_empoli"],),
                vm.NavLink(
                    label="Eleodora", icon="Robot 2", pages=["dados_gerais"],),
                vm.NavLink(
                    label="UrbanAI", icon="Globe", pages=["urbanai"]),
            ]
        )
    ),
)

Vizro().build(dashboard).run(debug=True)
```
